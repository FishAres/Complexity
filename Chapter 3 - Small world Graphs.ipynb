# Small-World Networks

Watts and Strogatz wanted to describe the small-world phenomenon demonstrated by Milgram. To do so, they created 

* a metric of **clustering** or "cliquishness" - likelihood two nodes are connected to the same node given they're connected to each other

* **path length** - the average distance between two nodes, i.e. degrees of separation

### Generative model for small-world graphs

1. Start with a **regular** graph of *n* nodes, each connected to *k* neighbors
    * A regular graph is one where each node is connected to the same number of neighbors (**degree** of the node)
    
    
2. Choose a subset of the edges and replace them with random edges
    * *p* is the probability that an edge is rewired
    
### Ring lattice (a regular graph):

```python
import networkx as nx
import matplotlib.pyplot as plt
# colors from our friends at http://colorbrewer2.org
COLORS = ['#8dd3c7','#ffffb3','#bebada','#fb8072','#80b1d3','#fdb462',
          '#b3de69','#fccde5','#d9d9d9','#bc80bd','#ccebc5','#ffed6f']

%matplotlib inline
```

```python
def adjacent_edges(nodes, halfk):
    n = len(nodes)
    for i, u in enumerate(nodes):
        for j in range(i+1, i+halfk+1):
            v = nodes[j % n] # wrapping
            yield u, v
            
def make_ring_lattice(n, k):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(adjacent_edges(nodes, k//2))
    return G
```

```python
lattice = make_ring_lattice(10, 4)
nx.draw_circular(lattice,
                node_color=COLORS[2],
                node_size=2000, with_labels=True)
plt.title("Ring lattice");
```
